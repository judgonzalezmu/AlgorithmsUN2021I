#5-1 Money Change Again

As we already know, a natural greedy strategy for the change problem does not work correctly for any set of
denominations. For example, if the available denominations are 1, 3, and 4, the greedy algorithm will change
6 cents using three coins (4 + 1 + 1) while it can be changed using just two coins (3 + 3). Your goal now is
to apply dynamic programming for solving the Money Change Problem for denominations 1, 3, and 4.

```python
%%writefile change_dp.cpp

#include <iostream>

int get_change(int m)
{
    int tip_mon[]={1,3,4};
    int mon_nec[m+1];
    mon_nec[0]=0;
    int a,mon;

    for(int i=0; i<=m; i++)
    {
        mon_nec[i+1]=1500;
        for(int j=0; j<3; j++)
        {
            a=tip_mon[j];
            if(i>=a)
            {
                mon= mon_nec[i-a]+1;
                if(mon<mon_nec[i])
                {
                    mon_nec[i]=mon;
                }
            }
        }
//        std::cout << mon_nec[i]<< " ";
    }
    return mon_nec[m];
}

int main() {
  int m;
  std::cin >> m;
  std::cout << get_change(m) << '\n';
}
```

#5-4 Longest Common Subsequence of Two Sequences
Compute the length of a longest common subsequence of two sequences.

```python
%%writefile lcs2.cpp

#include <iostream>
#include <vector>
#include <algorithm>

using std::vector;
using std::max;

int lcs2(vector<int> &a, vector<int> &b, int n, int m)
{
    int lcs[n+1][m+1]={};
    for(int i=1; i<=n; i++)
    {
        for(int j=1; j<=m; j++)
        {
            if(a[i-1]==b[j-1])
                lcs[i][j]=lcs[i-1][j-1]+1;
            else
                lcs[i][j] = max(lcs[i][j-1], max(lcs[i-1][j-1], lcs[i-1][j]));
        }
    }
    return lcs[n][m];
}

int main() {
  size_t n;
  std::cin >> n;
  vector<int> a(n);
  for (size_t i = 0; i < n; i++) {
    std::cin >> a[i];
  }

  size_t m;
  std::cin >> m;
  vector<int> b(m);
  for (size_t i = 0; i < m; i++) {
    std::cin >> b[i];
  }

  std::cout << lcs2(a, b, n, m) << std::endl;
}
```

#5-5 Longest Common Subsequence of Three Sequences
Compute the length of a longest common subsequence of three sequences.

```python
%%writefile lcs3.cpp

#include <iostream>
#include <vector>
#include <algorithm>

using std::vector;
using std::max;

int lcs3(vector<int> &a, vector<int> &b, vector<int> &c, int an, int bn, int cn)
{
    int lcs[cn+1][bn+1][an+1]={};

    for(int i=1; i<=an; i++)
    {
        for(int j=1; j<=bn; j++)
        {
            for(int k=1; k<=cn; k++)
            {
                if((a[i-1]==b[j-1])&&(b[j-1]==c[k-1]))
                    lcs[k][j][i] = lcs[k-1][j-1][i-1]+1;
                else
                    lcs[k][j][i] = max(lcs[k-1][j][i], max(lcs[k][j-1][i], lcs[k][j][i-1]));
            }
        }
    }
    return lcs[cn][bn][an];
}

int main() {
  size_t an;
  std::cin >> an;
  vector<int> a(an);
  for (size_t i = 0; i < an; i++) {
    std::cin >> a[i];
  }
  size_t bn;
  std::cin >> bn;
  vector<int> b(bn);
  for (size_t i = 0; i < bn; i++) {
    std::cin >> b[i];
  }
  size_t cn;
  std::cin >> cn;
  vector<int> c(cn);
  for (size_t i = 0; i < cn; i++) {
    std::cin >> c[i];
  }
  std::cout << lcs3(a, b, c, an, bn, cn) << std::endl;
}
```
