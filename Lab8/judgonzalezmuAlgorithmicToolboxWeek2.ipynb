#Problema 1

**Problem Description**\
**Task:** Given an integer $n^{th}$, find the $n^{th}$ Fibonacci number $F_n$.

**Input Format:** The input consists of a single integer $n$.

**Constraints:** $0 â‰¤ n â‰¤ 45$.

**Output Format:** Output $F_n$.

```python
%%writefile fibonacci.cpp
#include <iostream>
#include <cassert> 

int fibonacci_fast(int n) {
    int arr[n];
    arr[0]=0;
    arr[1]=1;

    for(int i=2; i<=n; i++)
    {
        arr[i]=arr[i-1]+arr[i-2];
    }

    return arr[n];
}

int main() {
    int n = 0;
    std::cin >> n;
    std::cout << fibonacci_fast(n) << '\n';
    return 0;
}
```

```python
!g++ fibonacci.cpp -o fibonacci
!./fibonacci
```

**Problem Description**

**Task:** Given an integer $n$, find the last digit of the $n^{th}$ Fibonacci number $F_n$ (that is, $F_n$ mod 10)

**Input Format:** The input consists of a single integer $n$

**Constraints:** $0 â‰¤ ð‘› â‰¤ 107$

**Output Format:** Output the last digit of $F_n$

```python
%%writefile fibonacci_last_digit.cpp
#include <iostream>

int get_fibonacci_last_digit_fast(int n) {
    int arr[n];
    arr[0]=0;
    arr[1]=1;

    for(int i=2; i<=n; i++)
    {
        arr[i]=(arr[i-1]+arr[i-2]) %10;
    }

    return arr[n];
}

int main() {
    int n;
    std::cin >> n;
    int c = get_fibonacci_last_digit_fast(n);
    std::cout << c << '\n';
    }
```

#Problema 3

**Problem Description**

**Task:** Given two integers $ð‘Ž$ and $ð‘$, find their greatest common divisor

**Input Format:** The two integers $ð‘Ž$, $ð‘$ are given in the same line separated by space

**Constraints:** $1 â‰¤ ð‘Ž$, $ð‘ â‰¤ 2 Â· 109$

**Output Format:** Output GCD($ð‘Ž$, $ð‘$)

```python
%%writefile gcd.cpp
#include <iostream>

int gcd_fast(int a, int b)
{
    int c=0;
    while(b!=0)
    {
        c=a%b;
        a=b;
        b=c;
    }
    return a;
}

int main() {
  int a, b;
  std::cin >> a >> b;  
  std::cout << gcd_fast(a, b) << std::endl;
  return 0;
}
```

#Problema 4

**Problem Description**

**Task:** Given two integers $ð‘Ž$ and $ð‘$, find their least common multiple

**Input Format:** The two integers $ð‘Ž$ and $ð‘$ are given in the same line separated by space

**Constraints:** $1 â‰¤ ð‘Ž$, $ð‘ â‰¤ 107$

**Output Format:** Output the least common multiple of $ð‘Ž$ and $ð‘$

```python
%%writefile lcm.cpp
#include <iostream>

int gcd(int a, int b)
{
    int c=0;
    while(b!=0)
    {
        c=a%b;
        a=b;
        b=c;
    }
    return a;
}

long long lcm_fast(int a, int b)
{
    long long d=(long long)a*b;
    long long e=gcd(a,b);
    long long f=d/e;
    return (long long) f;
}

int main() {
  int a, b;
  std::cin >> a >> b;
  std::cout << lcm_fast(a, b) << std::endl;
  return 0;
}
```
