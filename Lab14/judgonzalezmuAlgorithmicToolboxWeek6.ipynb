#6-1 Maximum Amount of Gold

You are given a set of bars of gold and your goal is to take as much gold as possible into
your bag. There is just one copy of each bar and for each bar you can either take it or not
(hence you cannot take a fraction of a bar).

```python
%%writefile knapsack.cpp

#include <iostream>
#include <vector>

using std::vector;

int optimal_weight(int W, const vector<int> &w, int n)
{
    int pe=0;
    vector<vector<int> > p(W+1,vector<int>(n+1));

    for(int i=1; i<=W; i++)
    {
        for(int j=1; j<=n; j++)
        {
            p[i][j]=p[i][j-1];
            if(w[j-1]<=i)
            {
                pe=p[i-w[j-1]][j-1]+w[j-1];
                if(pe>=p[i][j])
                    p[i][j]=pe;
            }
        }
    }
    return p[W][n];
}

int main() {
  int n, W;
  std::cin >> W >> n;
  vector<int> w(n);
  for (int i = 0; i < n; i++) {
    std::cin >> w[i];
  }
  std::cout << optimal_weight(W, w, n) << '\n';
}
```

#6-2 Partitioning Souvenirs
You and two of your friends have just returned back home after visiting various countries. Now you would
like to evenly split all the souvenirs that all three of you bought.

```python
%%writefile partition3.cpp

#include <iostream>
#include <vector>
#include <numeric>

using std::vector;

int partition3(vector<int> &A, int n)
{
    int total=std::accumulate(A.begin(),A.end(), 0);
    int w=0;
    int aux=0;
    int con=0;

    if(n<3)
        return 0;
    else if (total%3 != 0)
        return 0;
    else
    {
        w=total/3;
        vector<vector<int> > p(w+1,vector<int>(n+1));

        for(int i=1; i<=w; i++)
        {
            for(int j=1; j<=n; j++)
            {
                p[i][j]=p[i][j-1];
                if(A[j-1]<=i)
                {
                    aux=p[i-A[j-1]][j-1]+A[j-1];
                    if(aux>p[i][j])
                    {
                        p[i][j]=aux;
                    }
                }
                if(p[i][j]==w)
                {
                    con+=1;
                }
            }
        }
        if(con<3)
            return 0;
        else
            return 1;
    }

    return 0;
}

int main() {
  int n;
  std::cin >> n;
  vector<int> A(n);
  for (size_t i = 0; i < A.size(); ++i) {
    std::cin >> A[i];
  }
  std::cout << partition3(A, n) << '\n';
}
```
